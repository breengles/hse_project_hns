```python
from agent import Agent
from gym import make
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
def plot(log):
    fig, ax = plt.subplots(figsize=(12, 8))

    cfg = f'{log["alpha"][0]} | {log["gamma"][0]}'

    ax.set_title(f"Mean reward over 10 episodes: {cfg}")
    ax.set_xlabel("№ of transitions")
    ax.set_ylabel("Mean reward")

    plt.hlines(-110, np.min(log["step"]), np.max(log["step"]),
               colors="r", label="Solved")

    plt.plot(log["step"], log["mean"], label="Table Q")
    plt.fill_between(log["step"], log["mean"] - log["std"], log["mean"] + log["std"], alpha=0.5)
    plt.legend()
    plt.show()
```

```python
env = make("MountainCar-v0")
agent = Agent(env=env, alpha=0.15, gamma=0.98)
log = pd.DataFrame(agent.train())
plot(log)
```

```python
mean, std = agent.evaluate_policy(episodes=100)
print(f"Reward mean over 100 episodes: {mean} with std: {std}")
```

Видно, что агент в целом вышел на плато уже к 2 млн. взаимодействиям со средой.
